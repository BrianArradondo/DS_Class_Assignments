# Imports

```python
# imports
import pandas as pd
from sqlalchemy import create_engine
```

## Read in the Data as Pandas Dataframe

```python
school_df = pd.read_csv("School_Details.csv")
```

```python
school_df.head(10)
```

## Create a SQLite Database

- SQLite  is just one example of a database

```python
# create sqlite engine for a database called schools
engine = create_engine("sqlite:///schools.db", echo=True)
# create connection to engine
conn = engine.connect()
```

## Add a Table to the Database

```python
# add the dataframe as a table in sqlite
sqlite_table = "school_details"
school_df.to_sql(sqlite_table,conn,if_exists="fail")
```

```python
sql = "SELECT LOWESTCLASS FROM school_details Limit 5;"
pd.read_sql_query(sql,engine)
```

```python
sql = "SELECT `CRCDistance (in Km)` FROM school_details;"
pd.read_sql_query(sql,engine)
```

```python
sql = "SELECT EstablishmentName FROM school_details WHERE SchoolLevel = 'SSS';"

pd.read_sql_query(sql,engine)
```

```python
sql = "SELECT ManagementName From school_details WHERE SubdivisionName = 'Gangtok';"
pd.read_sql_query(sql,engine)
```
